```python
import matplotlib
import pandas as pd
from joblib import load
from trainer.data_loader_and_cleanser import DataLoader
%matplotlib inline
```

```python
def load_models():
    averaged_models_pkl = load("trainer/averaged_models.pkl")
    xgb_models_pkl = load("trainer/model_xgb.pkl")
    lgb_models_pkl = load("trainer/model_lgb.pkl")
    return averaged_models_pkl, xgb_models_pkl, lgb_models_pkl
```

```python
Averaged = 1 / 34.2450
XGBoost = 1 / 121.6862
LGBM = 1 / 172.3413
Sum = Averaged + XGBoost + LGBM
Averaged = Averaged / Sum
XGBoost = XGBoost / Sum
LGBM = LGBM / Sum
print(Averaged, XGBoost, LGBM)
```

```python
averaged_models_pkl, xgb_models_pkl, lgb_models_pkl = load_models()
```

### Show Microsoft Predictions

```python
df = DataLoader().prepare_test_data("MSFT", "files")
df_master = DataLoader().prepare_non_test_data("MSFT", "files")
```

```python
averaged_pred = averaged_models_pkl.predict(df.values)
xgb_pred = xgb_models_pkl.predict(df)
lgb_pred = lgb_models_pkl.predict(df)
```

```python
ensemble = averaged_pred * Averaged + xgb_pred * XGBoost + lgb_pred * LGBM
df["PercentChangeInRoe_pred"] = ensemble
df["ActualRoe"] = ((df["PercentChangeInRoe"] + 100.0) * df["PreviousRoe"]) / 100.0
df["ActualRoe_pred"] = ((df["PercentChangeInRoe_pred"] + 100.0) * df["PreviousRoe"]) / 100.0
df_master = df_master.set_index("PreviousRoe").join(df.set_index("PreviousRoe"))
df["ActualRoe"] = ((df["PercentChangeInRoe"] + 100.0) * df["PreviousRoe"]) / 100.0
```

```python
print(df_master.head())
```

```python
df_master = df_master[["Date", "ActualRoe", "ActualRoe_pred"]]
df_master["Date"] = pd.to_datetime(df_master["Date"], format="%m/%d/%Y")
df_master.set_index("Date", inplace=True)
```

```python
df_master.sort_values(by='Date')
```

```python
lines = df_master.plot.line()
```
